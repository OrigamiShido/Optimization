# 阻尼牛顿法

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def targetfunction(x,y):
    return 4*x**2+y**2-8*x-4*y

def grad(x,y):
    return np.array([8*x-8,2*y-4])

def hessian(x,y):
    return np.array([[8,0],[0,2]])

def damp_newton(x0,y0,epsilon):
    x = x0
    y = y0
    while True:
        g = grad(x,y)
        h = hessian(x,y)
        dk = -np.linalg.inv(h).dot(g) # newton方程的解
        if np.linalg.norm(dk)<epsilon:# 若解长小于阈值，则停止迭代
            break
        # armijo方法，用于对方向上进行线搜索
        alpha = 1
        while targetfunction(x+alpha*dk[0],y+alpha*dk[1])>targetfunction(x,y)+0.5*alpha*np.dot(g,dk):
            alpha = alpha/2
        x = x+alpha*dk[0]
        y = y+alpha*dk[1]
    return x,y

# 绘图
x=np.linspace(-10,10,500)
y=np.linspace(-10,10,500)
X,Y=np.meshgrid(x,y)
Z=targetfunction(X,Y)
plt.contour(X,Y,Z,20)
xt,yt=damp_newton(0,0,0.001)
plt.plot(xt,yt,'ro')
plt.title("Newton method")
plt.show()
print(f"x={xt},y={yt},f(x,y)={targetfunction(xt,yt)}")
```

# 牛顿-最速下降混合法

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math

def targetfunction(x,y):
    return 4*x**2+y**2-x**2*y

def grad(x,y):
    return np.array([8*x-2*y,2*y-x**2])

def hessian(x,y):
    return np.array([[8,-2],[-2,-2*x]])

def newton_steepest(x0,y0,epsilon):
    x = x0
    y = y0
    while True:
        g = grad(x,y)
        h = hessian(x,y)
        dk = -np.linalg.inv(h).dot(g)# newton方程的解
        if np.linalg.norm(dk)<epsilon:# 若解长小于阈值，则停止迭代
            break
        if np.linalg.det(h)==0:# 若矩阵奇异，则取负梯度方向
            dk = -g
        if np.dot(g,dk)>epsilon*np.linalg.norm(g)*np.linalg.norm(dk):# g负定，但逆存在，取反方向
            dk = -dk
        if abs(np.dot(g,dk))<=epsilon*np.linalg.norm(g)*np.linalg.norm(dk):# 当g和dk几乎正交时，采用负梯度方向
            dk = -g
        # armijo方法，用于对方向上进行线搜索
        alpha = 1
        while targetfunction(x+alpha*dk[0],y+alpha*dk[1])>targetfunction(x,y)+0.5*alpha*np.dot(g,dk):
            alpha = alpha/2
        x = x+alpha*dk[0]
        y = y+alpha*dk[1]
    return x,y
# 绘图
x=np.linspace(-10,10,500)
y=np.linspace(-10,10,500)
X,Y=np.meshgrid(x,y)
Z=targetfunction(X,Y)
plt.contour(X,Y,Z,50)
xt,yt=newton_steepest(0,0,0.001)
plt.plot(xt,yt,'ro')
plt.title("Newton-steepest method")
plt.show()
print(f"x={xt},y={yt},f(x,y)={targetfunction(xt,yt)}")
```
